```python
import nltk
from nltk.corpus import wordnet as wn

nltk.download('wordnet')
```

```python
wn.synsets('good')
wn.synsets('stupid')
wn.synsets('attack')
wn.synsets('protection')
```

```python
print(wn.synset('good.n.01').definition())
print(wn.synset('stupid.n.01').definition())
print(wn.synset('attack.n.02').definition())
print(wn.synset('protection.n.01').definition())
```

```python
print(wn.synset('good.n.01').examples())
print(wn.synset('stupid.n.01').examples())
print(wn.synset('attack.n.02').examples())
print(wn.synset('protection.n.01').examples())
```

```python
[str(lemma.name()) for lemma in wn.synset('good.n.01').lemmas()]
```
