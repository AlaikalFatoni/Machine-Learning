```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv('normalized_apple_prices.csv', header = None)
```

```python
df1.head()
```

```python
plt.plot(df1)
plt.xlabel('waktu periode (hari)')
plt.ylabel('normalized apple prices')
```

```python
dataset = df1.values
```

```python
def window_transform_series(series,window_size):
    X = []
    y = []
    for s in range(len(series)-window_size):
        X.append(series[s:s+window_size])
        y.append(series[s+window_size])
    X = np.asarray(X)
    X.shape = (np.shape(X)[0:2])
    y = np.asarray(y)
    y.shape = (len(y),1)
    return X,y
```

```python
X,y = window_transform_series(series = dataset,window_size = 7)
```

```python
split_data = int(np.ceil(2*len(y)/float(3)))
Xtrain = X[:split_data,:]
ytrain = y[:split_data]
```

```python
Xtest = X[split_data:,:]
ytest = y[split_data:]
```

```python
Xtrain = np.asarray(np.reshape(Xtrain, (Xtrain.shape[0],7,1)))
Xtest = np.asarray(np.reshape(Xtest, (Xtest.shape[0],7,1)))
```

```python
from keras.models import Sequential
from keras.layers import Dense, SimpleRNN
```

```python
rnn1 = Sequential()
rnn1.add(SimpleRNN(5, input_shape = (7,1)))
rnn1.add(Dense(1))
rnn1.compile(loss = "mean_squared_error")
print(rnn1.summary)
```

```python
hist = rnn1.fit(Xtrain, ytrain, epochs = 1000
               , batch_size = 500, verbose = 2
               , validation_split = 0.2)
```

```python
train.pred = rnn1.predict(Xtrain)
test.pred = rnn1.predict(Xtest)
print("training error: ",rnn1.evaluate(Xtrain, ytrain, verbose = 0))
print("testing error: ",rnn1.evaluate(Xtest, ytest, verbose = 0))
```
