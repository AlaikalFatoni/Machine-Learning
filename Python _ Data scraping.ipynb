```python
def url_to_transcript(url):
    page = requests.get(url).text
    soup = BeautifulSoup(page,'lxml')
    text = [p.text for p in soup.find_all('div', class_ = 'elementor_widget-theme-post-content')]
    print(url)
    return text
```
